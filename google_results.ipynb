```python
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup
```

```python
breweries = pd.read_csv('breweries.csv')
beers = pd.read_csv('beers.csv')
beers_simple = beers[['id', 'brewery_id', 'name', 'abv']]

beers_simple.head()
```

```python
#Rename brewery id column 
breweries_simple = breweries[['id', 'name']]
breweries_simple.rename(columns={'id':'brewery_id'}, inplace=True)
breweries_simple.head()
```

```python
merged = pd.merge(beers_simple, breweries_simple, how='inner', on='brewery_id')
merged.rename(columns={'name_x':'beer', 'name_y':'brewery'}, inplace=True)
merged.head()
```

```python
#Get brewery list
brew_list = []
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=True)
url = 'https://bigworldsmallgirl.com/central-texas-brewery-guide/#:~:text=With%20over%2058%20breweries%20in,love%20about%20our%20craft%20beer.'
browser.visit(url)
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
beer = soup.find('div', class_='entry-content clearfix')
brewerylist = beer.find_all('p')
b=[]
for p in brewerylist:
    if p.a:
        b.append(p.a)
brewery_list=[]
for a in b:
    if a.strong:
        print(a.strong.text)
        brew_list.append(a.strong.text)
        
```

```python
print(brew_list)
```

```python
#df[df['A'].isin([3, 6])]
filtered = merged[merged['brewery'].isin(brew_list)]
#filtered_df = merged[merged['brewery']=='Magic Hat']
filtered.head(20)
```

```python
brewery_google_dict = {}
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=True)
for brewery in brew_list:
    url = 'https://google.com/search?q=' + str(brewery)
    browser.visit(url)
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    result = soup.find('cite').text
    #links = result.find().text
    #print(result)
    brewery_google_dict[brewery] = result
#print(brewery_google_dict)
```

```python
#Make into df
results_df = pd.DataFrame(list(brewery_google_dict.items()),columns = ['Brewery','Link Result'])
results_df.to_csv('results.csv')

results_df
```
